```python
#using target guided ordinal encoding to replace the delivery_person_id it shows the efficiency of the delivery person
```

```python
def replace_deliveryperson_with_mean(df):
    mean_time = df.groupby('Delivery_person_ID')['Time_taken (min)'].mean().sort_values()
    df['Delivery_person_ID'] = df['Delivery_person_ID'].apply(lambda x: mean_time[x])
    return df

df = replace_deliveryperson_with_mean(df)
df.head()
```

```python
#working on latitude and longitude columns for both restaurant location and delivery location
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
#Restaurant location cluster
clus = df.loc[:,["Restaurant_latitude","Restaurant_longitude"]]
clus
```

```python
K_clusters = range(1,10)
kmeans = [KMeans(n_clusters=i) for i in K_clusters]
Y_axis = df[['Restaurant_latitude']]
X_axis = df[['Restaurant_longitude']]
score = [kmeans[i].fit(Y_axis).score(Y_axis) for i in range(len(kmeans))]
# Visualize
plt.plot(K_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init ='k-means++')
kmeans.fit(clus[clus.columns]) # Compute k-means clustering.
clus['Restaurant_cluster_label'] = kmeans.predict(clus[clus.columns])
```

```python
clus.head()
```

```python
clus["Restaurant_cluster_label"].unique()
```

```python
#delivery location cluster
d_clus = df.loc[:,["Delivery_location_latitude","Delivery_location_longitude"]]
d_clus
```

```python
K_clusters = range(1,10)
kmeans = [KMeans(n_clusters=i) for i in K_clusters]
Y_axis = df[['Delivery_location_latitude']]
X_axis = df[['Delivery_location_longitude']]
score = [kmeans[i].fit(Y_axis).score(Y_axis) for i in range(len(kmeans))]
# Visualize
plt.plot(K_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init ='k-means++')
kmeans.fit(d_clus[d_clus.columns]) # Compute k-means clustering.
d_clus['Delivery_cluster_label'] = kmeans.predict(d_clus[d_clus.columns])
```

```python
d_clus.head()
```

```python
df_cluster=pd.concat([clus["Restaurant_cluster_label"],d_clus["Delivery_cluster_label"]],axis=1)
```

```python
final_df = pd.concat([df,df_cluster],axis=1)
final_df.head()
```

```python
final_df.info()
```

```python
#As the restaurant location and delivery locations were clustered now we can drop lat and long columns
df = final_df.drop(labels=["Restaurant_latitude","Restaurant_longitude","Delivery_location_latitude","Delivery_location_longitude"],axis=1)
```
